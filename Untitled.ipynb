```python
#4. Codificación de variables categóricas
#antes de hacer el escalado, codificaremos las variables categoricas para que las entienda el modelo.
#las variables categoricas ordinales utilizaremos labelencoder y para las categoricas nominales usaremos onehotencoding
#variables categoricas ordinales son: Product_importance
#variables categoricas nominales son: Warehouse_block, Mode_of_Shipment, Gender
```

```python
# One-hot encoding aplicado a las columnas Warehouse_block, Mode_of_Shipment y Gender
#importamos las librerias OneHotEncoder y make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

df10 = df[['Warehouse_block', 'Mode_of_Shipment', 'Gender']]

transformer = make_column_transformer(
    (OneHotEncoder(), ['Warehouse_block', 'Mode_of_Shipment', 'Gender']),
    remainder='passthrough')

transformed = transformer.fit_transform(df10)
transformed_df = pd.DataFrame(transformed, columns=transformer.get_feature_names_out())
print(transformed_df.head())
transformed_df
```

```python
#unimos los dos dataframes que contienen las columnas que transformamos y las que habiamos separado
```

```python
df6 = pd.concat([transformed_df.reset_index(drop=True), df.reset_index(drop=True)], axis=1)
df6
```

```python
df6.isnull().sum()
#nose observan valores nulos 
```

```python
#eliminamos las columnas que no necesitamos por que las transformamos
df6.drop(['Warehouse_block', 'Mode_of_Shipment', 'Gender'], axis=1, inplace=True)

```

```python
#aplicar labelencoder para las variables categoricas ordinales
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
  

df6['Product_importance'] = label_encoder.fit_transform(df6['Product_importance'])
```

```python
#elegimos las variables a usar en el modelo
X = df6
Y = df6['Reached.on.Time_Y.N']
```

```python
#quitamos la columna etiqueta de la variable para usar de entrenamiento
X.drop(columns='Reached.on.Time_Y.N', axis=1, inplace=True)
```

```python
#entrenamos el modelo
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2, random_state=27)
```

```python
x_train.shape, x_test.shape
#revisamos la reparticion de los datos
```

```python
#usamos minmaxscaler por que standardscaler es para datos con distribucion normal
from sklearn.preprocessing import MinMaxScaler
mmsca = MinMaxScaler()
x_train = mmsca.fit_transform(x_train)
x_test = mmsca.fit_transform(x_test)
```

```python

# instanciamos un objeto con regresion logistica

clf = LogisticRegression()


clf.fit(X,Y)
```

```python
#prediccion del modelo
y_pred = clf.predict(x_train)



print('La accuracy score es:', accuracy_score(y_train, y_pred))
```

```python
from sklearn.metrics import recall_score
print('la sensibilidad(recall) es:', recall_score(y_train, y_pred))

```
